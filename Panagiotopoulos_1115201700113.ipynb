```python
import os 
import numpy as np
import pandas as pd
from PIL import Image
sets = {
    1 : (1,7),
    2 : (8,19),
    3 : (20,31),
    4 : (32,45),
    5 : (46,64)
}
path = "faces"
def loadimages(path,set_number):
    scaler = StandardScaler()
    start,finish = sets[set_number]
    X = []
    labels = []
    for i in range(1,11): #the loop for all the people 
        result1,rest1 = divmod(i,10)
        person = "person" + str(result1) + str(rest1) + "_"
        for j in range(start,finish+1): # the loop for the images of them
            result2,rest2 = divmod(j,10)
            imagename = path + '/' + person + str(result2) + str(rest2) + ".png"
            image = Image.open(imagename)
            image_vector = np.array(image).reshape(2500)
            X.append(image_vector)
            labels.append(i)
    X = np.squeeze(X)
    X = scaler.fit_transform(X)        
    X = np.array(X).T  # Transpose to have images as columns
    labels = np.array(labels)
    return X, labels         
```

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
def train_and_evaluate_eigenfaces(set_number, d):
    # Load images and labels for the specified set
    X, labels = loadimages(path, set_number)
    # Perform PCA with the specified number of components
    pca = PCA(n_components=d)
    pca.fit(X.T)
    # Project the training images onto the eigenspace
    projected_images = pca.transform(X.T)
    # Train a nearest neighbor classifier
    classifier = KNeighborsClassifier(n_neighbors=1)
    classifier.fit(projected_images, labels)
    accuracies = []
    # Evaluate the classification accuracy for each set
    for i in range(1, 6):
        test_data, test_labels = loadimages(path, i)
        # Project the test images onto the eigenspace
        projected_test_images = pca.transform(test_data.T)
        # Perform classification using the trained classifier
        predicted_labels = classifier.predict(projected_test_images)
        # Calculate the classification accuracy
        accuracy = accuracy_score(test_labels, predicted_labels)
        accuracies.append(accuracy)
    return accuracies

for i in range(1, 6):
    accuracies_d9 = train_and_evaluate_eigenfaces(i, d=9)
    accuracies_d30 = train_and_evaluate_eigenfaces(i, d=30)
    print("\nClassification Accuracies for", i)
    print("d = 9:", accuracies_d9)
    print("d = 30:", accuracies_d30)
```
